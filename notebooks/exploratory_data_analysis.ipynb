# **Load Libs**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

# **Env Configs**

```python
RAW_DATA = '../data/raw/creditcard.zip'
IMAGE_REPORT = '../reports/figures/'
```

# **Load Data**

```python
credit_transactions = pd.read_csv(RAW_DATA)

metadata =  pd.Series({
                "Memory usage": f"{credit_transactions.memory_usage().sum() / (1024 * 1024):.2f} MB",
                "Dataset shape": f"{credit_transactions.shape}"
            })

print(metadata)

credit_transactions.head()
```

## **Univariate Analysis**

```python
plt.figure(figsize=(10, 6))

plt.title(
    label='Amount Transaction Distribution',
    fontdict={'fontsize': 16, 'fontweight': 'bold'}
)

ax = sns.histplot(
    data=credit_transactions,
    x='Amount',
    bins=100,
    kde=False,
    color='steelblue'
)

plt.xlabel('Amount (â‚¬)')
plt.ylabel('Frequency (log scale)') 

plt.yscale('log')

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
ax.set_facecolor('#EAEAEA')

plt.tight_layout()

plt.savefig(f'{IMAGE_REPORT}amount-distribution.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
credit_transactions_processed = credit_transactions.copy()

credit_transactions_processed['Time'] = credit_transactions_processed['Time'].apply(lambda x: x // 3600 % 24)
credit_transactions_processed.rename(columns={'Time': 'Hour'}, inplace=True)


plt.figure(
    figsize=(10, 6),
    dpi=300
)

plt.title(
    label='Hourly Transaction Distribution',
    fontdict={'fontsize': 16, 'fontweight': 'bold'}
)

ax = sns.histplot(
    data=credit_transactions_processed,
    x='Hour',
    color='steelblue',
    bins=24
)

plt.xticks(np.arange(0, 25, 2))

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7, axis='y')
ax.set_facecolor('#EAEAEA')

plt.tight_layout()

plt.savefig(f'{IMAGE_REPORT}hourly-distribution.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
features = credit_transactions.columns.drop(['Time', 'Amount', 'Class'])

ncols = 4
nrows = np.ceil(len(features) // ncols)

fig, ax = plt.subplots(
    nrows=nrows,
    ncols=ncols,
    figsize=(4.5 * ncols, 4 * nrows)
)
    
ax = ax.flatten()

plt.suptitle(
    t="PCA Variables Distribution",
    fontsize=20,
    fontweight='bold'
)
    
for i, col in enumerate(features):
    sns.histplot(data=credit_transactions, x=col, ax=ax[i], color='steelblue',  label='Clean', kde=True)
    
    ax[i].set_title(f'{col}')
    ax[i].set_facecolor('#EAEAEA')
    
plt.tight_layout(rect=[0, 0, 1, 0.95])

fig.savefig(f"{IMAGE_REPORT}pca-transformd_variables-distribution.png", dpi=300, bbox_inches='tight')

plt.show()
```

## **Multivariate Analysis**

```python
scaler = StandardScaler()

credit_transactions_processed['Amount'] = scaler.fit_transform(credit_transactions_processed['Amount'].values.reshape(-1, 1))

clean = credit_transactions_processed['Class'] == 0
fraud = credit_transactions_processed['Class'] == 1

clean_transactions = credit_transactions_processed[clean]
fraud_transactions = credit_transactions_processed[fraud]

clean_transactions_sample = clean_transactions.sample(
    n=len(fraud_transactions),
    random_state=42
)

balanced_transactions = (
    pd.concat([clean_transactions_sample, fraud_transactions]))

balanced_transactions.reset_index(drop=True, inplace=True)

X = balanced_transactions.drop(columns=['Class', 'Hour'], axis=1)
y = balanced_transactions['Class']

tsne = TSNE(
    n_components=2,
    perplexity=25,
    random_state=42,
)

X_reduced_space = tsne.fit_transform(X)

df_reduced_space = pd.DataFrame(X_reduced_space, columns=['t-SNE 1', 't-SNE 2'])
df_reduced_space['Class'] = y.values
df_reduced_space['Class'] = df_reduced_space['Class'].map({0: 'Clean', 1: 'Fraud'})
df_reduced_space.head()
```

```python
plt.figure(figsize=(12, 8))

plt.title("t-SNE Distribution of Credit Card Transactions: Clean vs Fraud", fontsize=16, fontweight='bold', pad=20)
          
ax = sns.scatterplot(
    data=df_reduced_space,
    x='t-SNE 1', 
    y='t-SNE 2', 
    hue='Class', 
    palette={'Clean': 'steelblue', 'Fraud': 'firebrick'},
    alpha=0.7,
    s=50
)

plt.legend(
    title="Transaction",
    fontsize=10,
    title_fontsize=10,
    loc='upper right',
    frameon=True,
    shadow=True
)

plt.xlabel("t-SNE 1", fontsize=7)
plt.ylabel("t-SNE 2", fontsize=7)

plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
ax.set_facecolor('#EAEAEA')

plt.tight_layout()

plt.savefig(f"{IMAGE_REPORT}tsne-distribution-class.png", dpi=300, bbox_inches='tight')

plt.show()
```
